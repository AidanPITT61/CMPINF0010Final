# DATA

## Property Sales 

First part imports the required libraries, sets up the configuration, and loads the dataset into a DataFrame.
It also checks that the required columns exist.

```python
import pandas as pd
import matplotlib.pyplot as plt

# === Load Datasets ===
FILE_PATH = "PropertySales.csv"  
ZIP_COL = "PROPERTYZIP"
PRICE_COL = "PRICE"
SALE_DATE_COL = "SALEDATE"

# ---- Load dataset ----
df = pd.read_csv(FILE_PATH)

# Convert column data types
df[PRICE_COL] = pd.to_numeric(df[PRICE_COL], errors="coerce")
df[SALE_DATE_COL] = pd.to_datetime(df[SALE_DATE_COL], errors="coerce")

# Drop rows with missing ZIP codes or invalid prices
df = df.dropna(subset=[ZIP_COL, PRICE_COL])

print(f"Total records: {len(df)}")
df.head()
```

After, this section groups the data by ZIP code, counting the number of property sales and calculating both the total sales value and the average sale price for each area. It then sorts the results to find the ZIP codes with the highest number of property sales.

```python
summary = (
    df.groupby(ZIP_COL)
    .agg(
        Sales_Count=(PRICE_COL, "count"),           # Number of sales
        Total_Sales_Value=(PRICE_COL, "sum"),       # Total combined value
        Average_Sale_Price=(PRICE_COL, "mean"),     # Mean property price
    )
    .sort_values("Sales_Count", ascending=False)
    .reset_index()
)

# Round large values for better readability
summary["Total_Sales_Value"] = summary["Total_Sales_Value"].round(0).astype(int)
summary["Average_Sale_Price"] = summary["Average_Sale_Price"].round(0).astype(int)

# ---- Display the top 10 ZIP Codes ----
print("\nAreas with the Most Houses Sold (by ZIP Code):\n")
print(summary.head(10).to_string(index=False))
```

Finally, this section creates a bar graph to visualize which ZIP codes have the highest average property values among the top 10 areas by sales count. The Central Northside ZIP (15212) is highlighted in gold to show its strong performance in the housing market.

```python
# ---- Select the Top 10 ZIP Codes ----
top20 = summary.head(10)

# ---- Create Bar Chart ----
plt.figure(figsize=(12, 6))
bars = plt.barh(
    top20[ZIP_COL].astype(str),
    top20["Average_Sale_Price"],
    color="steelblue",
)

# ---- Highlight Central Northside ZIP ----
for i, zip_code in enumerate(top20[ZIP_COL]):
    if str(zip_code) == "15212":
        bars[i].set_color("gold")

# ---- Formatting ----
plt.xlabel("Average Sale Price ($)")
plt.ylabel("ZIP Code")
plt.title("Top 10 ZIP Codes by Property Sales and Average Sale Price")
plt.gca().invert_yaxis()  # Highest value at the top
plt.grid(axis="x", linestyle="--", linewidth=0.5, alpha=0.7)
plt.tight_layout()
plt.show()
```

## Public Transport
We start by importing our necessary libraries (pandas and matplotlib) and importing our data sets.

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

# === Load Datasets ===
FILE_PATH      = "AverageRiders.csv"   # Ridership data
OTP_FILE_PATH  = "OTP.csv"             # On-time performance data

# Read both CSVs
df  = pd.read_csv(FILE_PATH)
otp = pd.read_csv(OTP_FILE_PATH)
```

Before combining the data, we clean and reformat key columns:

```python
for col in ["route_full_name", "route"]:
    df[col]  = df[col].astype(str).str.strip()
    otp[col] = otp[col].astype(str).str.strip()

# Create route label (e.g., "61A — Wilkinsburg")
df["Route Label"]  = df["route"] + " — " + df["route_full_name"]
otp["Route Label"] = otp["route"] + " — " + otp["route_full_name"]

# Convert numerical fields
df["avg_riders"] = (
    df["avg_riders"]
    .astype(str)
    .str.replace(",", "", regex=False)
    .astype(float)
)
df["year_month"]  = pd.to_numeric(df["year_month"], errors="coerce")
otp["year_month"] = pd.to_numeric(otp["year_month"], errors="coerce")
otp["on_time_percent"] = pd.to_numeric(otp["on_time_percent"], errors="coerce")

# Drop incomplete rows
df  = df.dropna(subset=["avg_riders", "year_month"])
otp = otp.dropna(subset=["on_time_percent", "year_month"])
```

Next we filter for only the most recent avaliable year(2023 - 2024). This provides a more up to date and accurate measure.

```python
YEAR_START, YEAR_END = 2023, 2024

start_ym = YEAR_START * 100 + 1
end_ym   = YEAR_END * 100 + 12

# Filter data to our year range
df_range = df[(df["year_month"] >= start_ym) & (df["year_month"] <= end_ym)].copy()
otp_range = otp[(otp["year_month"] >= start_ym) & (otp["year_month"] <= end_ym)].copy()

# Average ridership per route
route_stats = (
    df_range.groupby(["Route Label", "route", "route_full_name"], as_index=False)["avg_riders"]
    .mean()
    .rename(columns={"avg_riders": f"Average Riders ({YEAR_START}-{YEAR_END})"})
)

# Average OTP per route
otp_avg = (
    otp_range.groupby(["Route Label", "route", "route_full_name"], as_index=False)["on_time_percent"]
    .mean()
    .rename(columns={"on_time_percent": f"OTP (Avg) {YEAR_START}-{YEAR_END}"})
)

# Merge both datasets
out = route_stats.merge(otp_avg, on=["Route Label", "route", "route_full_name"], how="left")

# Convert OTP to percentages
out[f"OTP (Avg) {YEAR_START}-{YEAR_END}"] = (out[f"OTP (Avg) {YEAR_START}-{YEAR_END}"] * 100).round(1)
```

Now we can rank the results. 

```python
TOP_N = 20
top_routes = (
    out.sort_values(f"Average Riders ({YEAR_START}-{YEAR_END})", ascending=False)
    .head(TOP_N)
    .reset_index(drop=True)
)
top_routes.insert(0, "Rank", range(1, len(top_routes) + 1))

display(top_routes)
```

Finally we build a visualization of the results.

```python
fig, ax = plt.subplots(figsize=(12, 6))

x = range(len(top_routes))
riders = top_routes[f"Average Riders ({YEAR_START}-{YEAR_END})"]
otp_pct = top_routes[f"OTP (Avg) {YEAR_START}-{YEAR_END}"]

# Bar chart (riders)
ax.bar(x, riders, color="steelblue", label="Average Riders")
ax.set_ylabel("Average Riders", color="steelblue")
ax.set_xlabel("Route")
ax.tick_params(axis="y", labelcolor="steelblue")

# Line (OTP)
ax2 = ax.twinx()
ax2.plot(x, otp_pct, color="darkorange", marker="o", linewidth=2, label="On-Time Performance (%)")
ax2.set_ylabel("On-Time Performance (%)", color="darkorange")
ax2.tick_params(axis="y", labelcolor="darkorange")

# X labels
ax.set_xticks(list(x))
ax.set_xticklabels(top_routes["Route Label"], rotation=45, ha="right")

# Title & grid
ax.set_title(f"Top {TOP_N} Routes: Ridership vs On-Time Performance ({YEAR_START}–{YEAR_END})")
ax.grid(True, axis="y", linestyle="--", linewidth=0.5)

# Legends
ax.legend(loc="upper left")
ax2.legend(loc="upper right")

plt.tight_layout()
plt.show()
```

# Arrest Data 

First we import and validate the data from the Arrest data set.

```python
import pandas as pd
import matplotlib.pyplot as plt

# === Load Data Sets ===
FILE_PATH = "ArrestDataSet.csv"
NEIGH_COL = "INCIDENTNEIGHBORHOOD"

df = pd.read_csv(FILE_PATH)

# Validate column
if NEIGH_COL not in df.columns:
    raise ValueError(
        f"Column '{NEIGH_COL}' not found. Available columns: {list(df.columns)}"
    )

print(f"Total records: {len(df)}")
df.head()
```

Next, we validate and clean the Neighborhood names then count arrest per neighborhood.

```python
# ---- Clean Neighborhood names ----
df[NEIGH_COL] = (
    df[NEIGH_COL]
    .astype(str)
    .str.strip()
    .replace({"": pd.NA, "nan": pd.NA, "None": pd.NA})
)

# ---- Remove invalid or out-of-region entries ----
exclude_terms = [
    "OUT OF COUNTY", 
    "OUT OF STATE", 
    "UNKNOWN", 
    "OTHER", 
    "UNSPECIFIED", 
    "AIRPORT"
]

# Keep only valid Pittsburgh neighborhoods
df = df[~df[NEIGH_COL].str.upper().isin(exclude_terms)]

# Drop missing values
df = df.dropna(subset=[NEIGH_COL])

# ---- Count arrests per neighborhood ----
counts = (
    df[NEIGH_COL]
    .value_counts()
    .rename_axis("Neighborhood")
    .reset_index(name="Count")
)

print(f"Unique neighborhoods remaining: {counts.shape[0]}")
counts.head()
```

Finally, we rank and print out the results.

```python
# ---- Sort and select 20 neighborhoods with the fewest arrests ----
lowest_20 = (
    counts.sort_values(["Count", "Neighborhood"], ascending=[True, True])
    .head(20)
    .reset_index(drop=True)
)

print("\nTop 20 Neighborhoods with the Fewest Arrests:\n")
print(lowest_20.to_string(index=False))

# ---- Highlight Central Northside ----
colors = [
    "gold" if "CENTRAL NORTHSIDE" in name.upper() else "seagreen"
    for name in lowest_20["Neighborhood"]
]

# ---- Create Bar Chart ----
plt.figure(figsize=(10, 6))
plt.barh(lowest_20["Neighborhood"], lowest_20["Count"], color=colors)
plt.xlabel("Number of Arrests")
plt.ylabel("Neighborhood")
plt.title("Top 20 Neighborhoods with the Fewest Arrests in Pittsburgh (Valid City Data Only)")
plt.gca().invert_yaxis()  # Lowest crime at top
plt.grid(axis="x", linestyle="--", linewidth=0.5, alpha=0.7)
plt.tight_layout()
plt.show()
```
