```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('PropSalesDataSet.csv')
# ensure price numeric
df['PRICE'] = pd.to_numeric(df['PRICE'], errors='coerce')

# group by school district
group = df.groupby('SCHOOLDESC').agg(
    Sale_Count=('PRICE','size'),
    Total_Sales_Value=('PRICE','sum')
).sort_values('Sale_Count', ascending=False)

# top 15
top15 = group.head(15)
# least 15
least15 = group.tail(15)

# function to plot side-by-side bars
def plot_side_by_side(data, title):
    x = np.arange(len(data))
    width = 0.4
    plt.figure(figsize=(25,10))
    plt.bar(x - width/2, data['Sale_Count'], width=width)
    plt.bar(x + width/2, data['Total_Sales_Value'], width=width)
    plt.xticks(x, data.index, rotation=45, ha='right')
    plt.xlabel("School District")
    plt.ylabel("Value")
    plt.title(title)
    plt.legend(["Sale Count", "Total Sales Value"])
    plt.tight_layout()
    plt.show()

plot_side_by_side(top15, "Top 15 School Districts by Property Sales")
plot_side_by_side(least15, "Least 15 School Districts by Property Sales")
```
