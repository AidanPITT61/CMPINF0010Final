This code reads and organizes the arrest dataset. It identifies the top 15 neighborhoods with the highest number of arrests by counting how often each neighborhood is mentioned. It also identifies the 15 neighborhoods with the lowest number of arrests. After collecting the data, the program organizes it into a readable list that shows each neighborhood’s arrest count and rank. This information provides an estimate of the overall safety of Pittsburgh neighborhoods. The input and output parameters can also be adjusted to display more than 15 results, such as the top 20, 30, or 50 neighborhoods.

```python
import pandas as pd

FILE_PATH = "ArrestDataSet.csv"
NEIGH_COL = "INCIDENTNEIGHBORHOOD"

def main():
    # Load Arrest Data Set
    df = pd.read_csv(FILE_PATH)

    # Validate column
    if NEIGH_COL not in df.columns:
        raise ValueError(
            f"Column '{NEIGH_COL}' not found. Available columns: {list(df.columns)}"
        )

    # Clean neighborhood values: strip spaces and standardize missing arrest
    df[NEIGH_COL] = (
        df[NEIGH_COL]
        .astype(str)
        .str.strip()
        .replace({"": pd.NA, "nan": pd.NA, "None": pd.NA})
    )

    # Count arrests per neighborhood
    counts = (
        df[NEIGH_COL]
        .dropna()
        .value_counts()
        .rename_axis("Neighborhood")
        .reset_index(name="Count")
    )

    # Top 15 highest and lowest
    top15_highest = (
        counts.sort_values(["Count", "Neighborhood"], ascending=[False, True])
        .head(15)
        .reset_index(drop=True)
    )
    top15_lowest = (
        counts.sort_values(["Count", "Neighborhood"], ascending=[True, True])
        .head(15)
        .reset_index(drop=True)
    )

    # Show results in console
    print("\nTop 15 Highest Arrest Neighborhoods:\n") # Display Top 15 Highest
    print(top15_highest.to_string(index=False))

    print("\nTop 15 Lowest Arrest Neighborhoods:\n") # Display Top 15 Lowest
    print(top15_lowest.to_string(index=False))

if __name__ == "__main__":
    main()
```

In conclusion, the data suggests that smaller neighborhoods and those located farther from the city tend to have fewer crimes and arrests, while neighborhoods closer to downtown Pittsburgh experience higher arrest rates. This pattern may also be influenced by population differences, as neighborhoods with fewer residents naturally record fewer arrests. Therefore, the data indicates that proximity to the city and population size are correlated with arrest frequency, which means the closer and more populated a neighborhood is, the higher the number of arrests is likely to be.
